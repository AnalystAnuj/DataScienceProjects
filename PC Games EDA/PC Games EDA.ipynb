```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Top_10_Developers= df['Developer'].value_counts().head(10)
Top_10_Developers 
```

```python
#top 10 game developing companies
plt.style.use("ggplot")
plt.figure(figsize=(15,10))
Top_10_Developers.plot(kind="barh",color="#1F8BC1",title="Top 10 Computer Games Developing Companies")
plt.show()
```

From above graph
1. Maxis is leading developer and
2. TalonSoft is least developer

```python
#Now let's analyze the top 10 producing companies.
Top_10_producers = df['Producer'].value_counts().head(10)
Top_10_producers
```

```python
#top 10 game producing companies
plt.style.use("bmh")
plt.figure(figsize=(15,10))
Top_10_producers.plot(kind="barh",color="#1FC1B7",title="Top 10 Computer Games Producing Companies")
plt.show()
```

As can be seen from the graph, Electronic Arts company leads the list.

```python
# Now let's take a look at the game types
Top_10_game_genres =df["Genre"].value_counts().head(10)
Top_10_game_genres
```

```python
# Top 10 Computer Game Genres
plt.style.use("bmh")
plt.figure(figsize=(15,10))
Top_10_game_genres.plot(kind="barh",color="#E8051F",title="Top 10 Computer Game Genres")
plt.show()
```

As can be seen from the graphic, 103 games are in the FPS type. Fps genre game is followed by RTS(48) AND Action-adventure (42) game genres.

```python
#Let's take a look at operating systems.
OS_of_Games = df["Operating System"].value_counts().head(10)
OS_of_Games
```

```python
# Top 10 Computer Game Operating System
plt.style.use("bmh")
plt.figure(figsize=(15,10))
OS_of_Games.plot(kind="barh",color="#F55347",title="Top 10 Computer Game Operating System")
plt.show()
```

As can be seen from the graph, the Microsoft Windows operating system is at the top of the list.
